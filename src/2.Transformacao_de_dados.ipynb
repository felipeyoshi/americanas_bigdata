#Transformação de Dados - Construção da camada Silver

##1. Carregar os arquivos JSON da camada Bronze em Spark Dataframe

```
MOUNT_RAW = '/mnt/databricks-felipeyoshimoto/bronze/iota/'

# Carregar os arquivos JSON pelo spark.read e alocar em Spark Dataframe
df_indicadores = spark.read.json(MOUNT_RAW + 'indicadores.json')
df_variaveis = spark.read.json(MOUNT_RAW + 'variaveis.json')
```

##2. Transformar os dados semi-estruturados em estruturados (processo realizado em memória)

```
from pyspark.sql.functions import explode

df_indicadores_cleaned = (
  df_indicadores.select(explode("data").alias("data_exp")).select(
    "data_exp.axis_aux1",
    "data_exp.axis_aux2",
    "data_exp.axis_aux3",
    "data_exp.axis_name",
    "data_exp.city_id",
    "data_exp.city_name",
    "data_exp.explanation",
    "data_exp.formula",
    "data_exp.formula_human",
    "data_exp.goal",
    "data_exp.goal_explanation",
    "data_exp.goal_operator",
    "data_exp.goal_source",
    "data_exp.indicator_id",
    "data_exp.indicator_name",
    "data_exp.institute_id",
    "data_exp.justification_of_missing_field",
    "data_exp.observations",
    "data_exp.period",
    "data_exp.region_dl",
    "data_exp.region_id",
    "data_exp.region_name",
    "data_exp.state_name",
    "data_exp.state_uf",
    "data_exp.tags",
    "data_exp.technical_information",
    "data_exp.updated_at",
    "data_exp.user_goal",
    "data_exp.user_id",
    "data_exp.valid_from",
    "data_exp.value",
    "data_exp.variation_name",
    "data_exp.variation_order"
  )
)
```

```
display(df_indicadores_cleaned)
```

```
from pyspark.sql.functions import explode

df_variaveis_cleaned = (
  df_variaveis.select(explode("data").alias("data_exp")).select(
    "data_exp.city_id",
    "data_exp.city_name",
    "data_exp.cognomen",
    "data_exp.exp_source",
    "data_exp.institute_id",
    "data_exp.is_basic",
    "data_exp.measurement_unit_name",
    "data_exp.name",
    "data_exp.observations",
    "data_exp.period",
    "data_exp.source",
    "data_exp.type",
    "data_exp.updated_at",
    "data_exp.user_id",
    "data_exp.valid_from",
    "data_exp.value",
    "data_exp.variable_id"
  )
)
```

```
display(df_variaveis_cleaned)
```

##3. Salvar os dataframes transformados na camada Silver

```
MOUNT_CLEANED_INDICADORES = '/mnt/databricks-felipeyoshimoto/silver/iota/indicadores/'
MOUNT_CLEANED_VARIAVEIS = '/mnt/databricks-felipeyoshimoto/silver/iota/variaveis/'

df_indicadores_cleaned.write.mode("overwrite").parquet(MOUNT_CLEANED_INDICADORES)
df_variaveis_cleaned.write.mode("overwrite").parquet(MOUNT_CLEANED_VARIAVEIS)
```
