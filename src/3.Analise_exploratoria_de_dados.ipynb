###1. Análise Exploratória

###1.1. Agrupamento de dados - Indicadores

```
%sql

CREATE OR REPLACE TEMP VIEW VIEW_TOP5_INDICADORES AS (  
  SELECT 
    indicator_id,
    indicator_name,
    COUNT(1) count_indicators
  FROM VIEW_INDICADORES
  WHERE value is not null and value <> 0
  GROUP BY 1, 2
  ORDER BY 3 DESC
  LIMIT 5
);
```

```
%sql

SELECT * FROM VIEW_TOP5_INDICADORES;
```

###1.2. Agrupamento de dados - Variáveis

```
%sql

CREATE OR REPLACE TEMP VIEW VIEW_TOP5_VARIAVEIS AS ( 
  SELECT
    variable_id,
    cognomen,
    COUNT(1) count_variaveis
  FROM VIEW_VARIAVEIS
  WHERE value is not null and value <> 0
  GROUP BY 1, 2
  ORDER BY 3 DESC
  LIMIT 5
);
```

```
%sql

SELECT * FROM VIEW_TOP5_VARIAVEIS;
```

###1.3. Seleciona o lançamento mais atualizado para os 3 indicadores com maior número de registros

```
%sql

CREATE OR REPLACE TEMP VIEW VIEW_INDICADOR_POPULACAO_TOTAL AS (
  WITH INDICADORES_RANKEADOS AS (
    SELECT 
      city_id,
      city_name,
      state_uf,
      state_name,
      indicator_id,
      indicator_name,
      value,
      updated_at,
      DENSE_RANK() OVER (PARTITION BY city_id, state_uf, indicator_id ORDER BY updated_at, value DESC) AS RANKING
    FROM VIEW_INDICADORES
    WHERE indicator_id = 361 and value is not null and value <> 0
  ),

  ULTIMO_INDICADOR AS (
    SELECT
      *
    FROM INDICADORES_RANKEADOS
    WHERE RANKING = 1
  )

  SELECT DISTINCT
    *
  FROM ULTIMO_INDICADOR
)
```

```
%sql

CREATE OR REPLACE TEMP VIEW VIEW_INDICADOR_TAXA_DE_HOMICIDIO AS (
  WITH INDICADORES_RANKEADOS AS (
    SELECT 
      city_id,
      city_name,
      state_uf,
      state_name,
      indicator_id,
      indicator_name,
      value,
      updated_at,
      DENSE_RANK() OVER (PARTITION BY city_id, state_uf, indicator_id ORDER BY updated_at, value DESC) AS RANKING
    FROM VIEW_INDICADORES
    WHERE indicator_id = 534 and value is not null and value <> 0
  ),

  ULTIMO_INDICADOR AS (
    SELECT
      *
    FROM INDICADORES_RANKEADOS
    WHERE RANKING = 1
  )

  SELECT DISTINCT
    *
  FROM ULTIMO_INDICADOR
)
```

```
%sql

CREATE OR REPLACE TEMP VIEW VIEW_INDICADOR_MORTES_NO_TRANSITO AS (
  WITH INDICADORES_RANKEADOS AS (
    SELECT 
      city_id,
      city_name,
      state_uf,
      state_name,
      indicator_id,
      indicator_name,
      value,
      updated_at,
      DENSE_RANK() OVER (PARTITION BY city_id, state_uf, indicator_id ORDER BY updated_at, value DESC) AS RANKING
    FROM VIEW_INDICADORES
    WHERE indicator_id = 4044 and value is not null and value <> 0
  ),

  ULTIMO_INDICADOR AS (
    SELECT
      *
    FROM INDICADORES_RANKEADOS
    WHERE RANKING = 1
  )

  SELECT DISTINCT
    *
  FROM ULTIMO_INDICADOR
)
```

###2. Persistência dos dados na camada Gold

###2.1. Cria as tabelas na camada Gold para o lançamento mais atualizado dos 3 indicadores com maior número de registros

```
MOUNT_CURATED_INDICADORES = '/mnt/databricks-felipeyoshimoto/gold/iota/indicadores/'

spark.sql('SELECT * FROM VIEW_INDICADOR_POPULACAO_TOTAL').write.mode('overwrite').parquet(MOUNT_CURATED_INDICADORES + 'ultimo_indicador_populacao_total')
spark.sql('SELECT * FROM VIEW_INDICADOR_TAXA_DE_HOMICIDIO').write.mode('overwrite').parquet(MOUNT_CURATED_INDICADORES + 'ultimo_indicador_taxa_de_homicidio')
spark.sql('SELECT * FROM VIEW_INDICADOR_MORTES_NO_TRANSITO').write.mode('overwrite').parquet(MOUNT_CURATED_INDICADORES + 'ultimo_indicador_mortes_no_transito')
```

###2.2. Cria as tabelas na camada Gold para o lançamento mais atualizado das 3 variáveis com maior número de registros

```
MOUNT_CURATED_VARIAVEIS = '/mnt/databricks-felipeyoshimoto/gold/iota/variaveis/'

spark.sql('SELECT * FROM VIEW_VARIAVEIS_POPULACAO_TOTAL').write.mode('overwrite').parquet(MOUNT_CURATED_VARIAVEIS + 'ultima_variavel_populacao_total')
spark.sql('SELECT * FROM VIEW_VARIAVEIS_HOMICIDIOS').write.mode('overwrite').parquet(MOUNT_CURATED_VARIAVEIS + 'ultima_variavel_homicidios')
spark.sql('SELECT * FROM VIEW_VARIAVEIS_HOMENS').write.mode('overwrite').parquet(MOUNT_CURATED_VARIAVEIS + 'ultima_variavel_populacao_de_homens')
```
